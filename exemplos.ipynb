```python
import pickle
import pandas as pd
def load_from_pickle(directory):
    return pickle.load(open(directory,"rb"))

# Loading the data
data = load_from_pickle(directory="merged_training.pkl")
sliced_data = data[:20]



emoroberta = pipeline('sentiment-analysis', model='arpanghoshal/EmoRoBERTa')
# print(f'EmoRoberta: {emotion(text)}')
# sliced_data.assign(emoroberta=lambda df : emotion(str(df.text)))
sliced_data['emoroberta'] = sliced_data['text'].apply(lambda x: emoroberta(x))


emotion_classifier = pipeline("sentiment-analysis", model="michellejieli/emotion_text_classifier")
# print(f'EmotionClassifier: {emotion_classifier(text)}')
sliced_data['emotion_classifier'] = sliced_data['text'].apply(lambda x: emotion_classifier(x))

distilbert_uncased = pipeline("text-classification", model="bhadresh-savani/distilbert-base-uncased-emotion")
# print(f'distilbert_uncased: {distilbert_uncased(text)}')
sliced_data['distilbert_uncased'] = sliced_data['text'].apply(lambda x: distilbert_uncased(x))

bert_based = pipeline("text-classification", model="bhadresh-savani/bert-base-uncased-emotion")
# print(f'bert_based: {bert_based(text)}')
sliced_data['bert_based'] = sliced_data['text'].apply(lambda x: bert_based(x))

emotion_english = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")
# print(f'emotion_english: {emotion_english(text)}')
sliced_data['emotion_english'] = sliced_data['text'].apply(lambda x: emotion_english(x))

with pd.option_context('display.max_colwidth', None):
  display(sliced_data)
```
